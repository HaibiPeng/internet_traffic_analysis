```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import math
import seaborn as sns
import statsmodels.api as sm
```

```python
flowsdata=[]
with open("/Users/jdszsl/Desktop/Aalto/Aalto-Communications Engineering/ELEC-E7130 - Internet Traffic Measurements and Analysis/Assignments/Assignment4/data/flows.txt", "r") as f:  # 
    for line in f.readlines():
        line = line.strip('\n') #get rid of '\n'
        flowsdata.append(int(line)) #str to int
print(flowsdata)
```

```python
#Scatterplot
xValue = list(range(1, len(flowsdata)+1))
yValue = [x for x in flowsdata]

plt.figure(figsize=(16,6))
plt.title('Scatterplot')
plt.xlabel('Number of observation')
plt.ylabel('Values of flow lengths in bytes')

plt.scatter(xValue, yValue, s=20, marker='o')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(16,6))
plt.title('Histogram')
plt.xlabel('Values of flow lengths in bytes')
plt.ylabel('Number of observation')
#bins = np.linspace(min(yValue),max(yValue),1000)
plt.hist([x for x in flowsdata], bins=100, log=1, histtype='bar', rwidth=0.8)
```

```python
#Boxplot
plt.figure(figsize=(16,6))
plt.title('Boxplot')
plt.xlabel('Values of flow lengths in bytes')
plt.ylabel('flows')

plt.boxplot(x = [x for x in flowsdata],
            patch_artist=True,
            showmeans=True,
            showfliers=True,
            vert=False,
            boxprops = {'color':'black','facecolor':'#9999ff'},
            flierprops = {'marker':'o','markerfacecolor':'red','color':'black'},
            meanprops = {'marker':'D','markerfacecolor':'indianred'},
            medianprops = {'linestyle':'--','color':'orange'})

#plt.xlim(0,1000)
plt.show()
```

```python
#Empirical CDF
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)
    # x-data for the ECDF: x
    x = np.sort(data)
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y

plt.figure(figsize=(16,6))
x = [x for x in flowsdata]
x_std1,y_std1=ecdf(x)
kwargs = {'cumulative': True, 'density': True}
sns.distplot(x, hist_kws=kwargs, kde_kws={'cumulative': True})
#plt.plot(x_std1,y_std1,marker='.',markersize=5.,linestyle='none')
plt.title('Empirical CDF')
plt.xlabel('Values of flow lengths in bytes')
```
