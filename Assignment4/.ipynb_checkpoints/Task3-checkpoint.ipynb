```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
import matplotlib
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot
```

```python
linkdata=[]
for i in range(1,5):
    data=[]
    with open("/Users/jdszsl/Desktop/Aalto/Aalto-CommunicationsEngineering/ELEC-E7130 - Internet Traffic Measurements and Analysis/Assignments/Assignment4/data/linkload-"+str(i)+".txt", "r") as f:  # 打开文件
        for line in f.readlines():
            line = line.strip('\n') #get rid of '\n'
            if i == 2:
                data.append(float(line.split(' ')[0])) #str to float
            else:
                data.append(float(line.split(' ')[1])) #str to float
        linkdata.append(data)
        data=[]
```

```python
print(linkdata[3])
```

```python
#lag plot
for i in range(0,4):
    s = pd.Series(linkdata[i])
    plt.figure(figsize=(16,6))
    lag_plot(s, lag=1)
    plt.title('Lag plot of link'+str(i+1))
    plt.show()
```

```python
#correlogram
for i in range(0,4):
    s = pd.Series(linkdata[i])
    plt.figure(figsize=(16,6))
    autocorrelation_plot(s)
    plt.title('Correlogram of link'+str(i+1))
    plt.show()
```
