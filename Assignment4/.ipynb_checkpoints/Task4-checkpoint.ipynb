```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import math
import seaborn as sns
import time
```

```python
csv_data = pd.read_csv('/Users/jdszsl/Desktop/Aalto/Aalto-Communications Engineering/ELEC-E7130 - Internet Traffic Measurements and Analysis/Assignments/Assignment4/data/bytes.csv')
print(csv_data)
```

```python
RXbytes=csv_data.iloc[:,1].values
RXpackets=csv_data.iloc[:,2].values
TXbytes=csv_data.iloc[:,3].values
TXpackets=csv_data.iloc[:,4].values
```

```python
T1 = time.time()
sns.set_style('white')
p = sns.pairplot(csv_data, kind = 'scatter', diag_kind = 'hist', palette = 'husl')
p.map_upper(sns.scatterplot,color='red')
p.map_lower(sns.scatterplot, color='green')
p.map_diag(plt.hist)
T2 = time.time()
print('data handling takes:%ss' % (T2 - T1))
```
